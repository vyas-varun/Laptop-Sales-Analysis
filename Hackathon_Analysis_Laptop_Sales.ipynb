```python
Sales_sub = Sales[['StorePostcode','RetailPrice']]
Sales_sub['AverageRetailprice'] = Sales_sub.groupby('StorePostcode')['RetailPrice'].transform(lambda x: x.mean())

Sales_df = Sales_sub.drop('RetailPrice',axis=1).drop_duplicates().sort_values(by='AverageRetailprice',ascending=False).head(20)
Sales_df
```

```python
plt.figure(figsize=(12,8))
plt.rcParams['font.size'] = '15'
plt.bar(x=Sales_df['StorePostcode'],height=Sales_df['AverageRetailprice'], width=0.4 , align='center');
plt.xticks(rotation=-90);
plt.ylabel('Average Retail Price');
plt.title("Average Retail Price by Store")
```

```python
sns.set(rc={'figure.figsize':(10,6)})
sns.boxplot(y='StorePostcode', x='RetailPrice',data=Sales);
plt.title("Compare retail prices across stores, side-by-side boxplots of retail price by store.")
```

```python
plt.figure(figsize=(10,10))
plt.rcParams['font.size'] = '12'

colors = sns.color_palette('bright')[0:4]

plt.pie(x=Sales_df['AverageRetailprice'],
        labels=Sales_df['StorePostcode'], 
        colors  = colors,
        wedgeprops = { 'linewidth' : 7, 'edgecolor' : 'white' },
        autopct='%.0f%%')
circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(circle)

plt.title("Average Retail Price of Laptop Stores Post Code")
plt.show()
```

```python
Sales_subset =  Sales[['OS X Customer','OS Y Customer','OS X Store','OS Y Store']]

corrs= Sales_subset.corr()
sns.heatmap(corrs,annot=True,cmap=plt.cm.Blues);

plt.title("Comparison of OS Variables")
```

```python
sns.set(rc={'figure.figsize':(15,8)})
sns.set_theme(style='darkgrid')
sns.lineplot(x='ProcessorSpeeds',y='Configuration',data=Sales,markers=True)
plt.xticks(rotation=-45);
plt.title("Comparison of Processors Speed with the configuration of laptops")
```

```python
sns.set_theme(style="darkgrid")

g = sns.catplot(
    data = Sales, 
    kind="bar",
    x="Date", 
    y="RetailPrice", hue="Battery Life (Hours)",
    palette="Reds", alpha=.9, height = 8
)
g.set_axis_labels("Date", "Retail Price")
g.legend.set_title("Battery Life")

plt.title("Daily Retail Price comparison with respect to Battery Life")
```

```python
Bundle_y['RetailPrice_diff'] = Bundle_y['Retail Price'] - Bundle_y['Retail Price'].pct_change(periods=1) 
Bundle_n['RetailPrice_diff'] = Bundle_n['Retail Price'] - Bundle_n['Retail Price'].pct_change(periods=1)
```

```python
plt.figure(figsize=(20,8))
sns.set_theme(style='darkgrid')
sns.lineplot(x='Date',y='RetailPrice_diff',data=Bundle_y,color='skyblue');
sns.lineplot(x='Date',y='RetailPrice_diff',data=Bundle_n,color='pink');
plt.title("Visualizing Retail Price between if application is a bundle or not")
```

```python
sns.set_theme(style='darkgrid')
sns.lmplot(data=Sales,x='HD Size (GB)',y='RetailPrice',hue='RAM (GB)',height=8);
plt.title("Retail Price VS Display(HD Size) with respect to RAM configration")
```

```python
sns.set_theme(style='darkgrid')
sns.jointplot(data=Sales,x='CustomerStoreDistance',y='RetailPrice',height=10,hue='Integrated Wireless?');
plt.title("Comparison of Store distance with Retail Price filtered with Integrated Wireless")
```
